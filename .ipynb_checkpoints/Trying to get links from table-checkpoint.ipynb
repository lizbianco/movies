```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import re
import pandas as pd
```

```python
#sets starting point
url = 'http://www.boxofficemojo.com/franchises/'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page, 'lxml')
```

```python
all_links =[]
franchise_links=[]

for link in soup.find_all('a'):
    all_links.append(link.get('href'))

```

```python
for item in all_links:
    if item[0]=='.':
        franchise_links.append(item)
```

```python
movie_links=[]
for item in franchise_links:
    url = 'http://www.boxofficemojo.com/franchises'+item[1:]
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page, 'lxml')
    table = soup.find_all('table')[2]
    for link in table.find_all('a'):
        i = link.get('href')
        if i.startswith('/movie') and i not in movie_links:
            movie_links.append(i)
            
for i in movie_links:
    print(i)

    

```
